#### Analisando os Outliers

```python
df_netflix.info()
```

```python
df_netflix['Viewership Score'].plot.box()
```

```python
df_netflix['Viewership Score'].plot.hist()
```

```python
df_netflix.plot.box(column='Viewership Score', by='Rank')
```

```python
df_netflix.describe()
```

```python
q1_viewship = df_netflix['Viewership Score'].quantile(0.25)
q3_viewship = df_netflix['Viewership Score'].quantile(0.75)
iqr_viewship = q3_viewship - q1_viewship
iqr_viewship
```

```python
limite_inferior = q1_viewship - (1.5 * iqr_viewship)
limite_superior = q3_viewship + (1.5 * iqr_viewship)
limite_inferior, limite_superior
```

```python
df_netflix['Viewership Score'][(df_netflix['Viewership Score'] < limite_inferior) | (df_netflix['Viewership Score'] > limite_superior)]
```

698 Outliers

```python
# Zscores
from scipy.stats import zscore
import numpy as np
```

```python
z = np.abs(zscore(df_netflix['Viewership Score']))
```

```python
z
```

```python
df_netflix['Viewership Score'][z > 3.0]
```
